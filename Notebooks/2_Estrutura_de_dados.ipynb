###Estrutura de dados em Python

####**Listas**

```
# Criando uma lista 
animais = [1,2,3]
animais
```

```
animais = ["cachorro", "gato", 12345, 6.5]
animais
```

```
# Primeiro elemento da lista
animais[0]
```

```
# Quarto elemento da lista
animais[3]
```

```
# Substituir o primeiro elemento da lista
animais[0] = "papagaio"
```

```
animais
```

```
# Remover gato da lista
animais.remove("gato")
```

```
animais
```

```
# Tamanho da lista
len(animais)
```

```
# Verificar se gato está na lista
"gato" in animais
```

```
lista = [500, 30, 300, 80, 10]
```

```
# Valor máximo
max(lista)
```

```
# Valor mínimo
min(lista)
```

```
# Criar uma lista aninhada
animais.append(["leão", "Cachorro"])
```

```
animais
```

```
# Adicionar elementos à uma lista
animais.extend(["cobra", 6])
```

```
animais
```

```
# Contar a palavra leão na lista
animais.count("leão")
```

```
# Ordenar a lista
lista.sort()
```

```
lista
```

####**Tuplas**

```
# As tuplas usam parênteses como síntaxe
tp = ("Banana", "Maçã", 10, 50)
```

```
# Primeiro elemento
tp[0]
```

```
# Diferente das listas as tuplas são imutáveis, o que quer dizer que não podemos alterar os seus elementos
tp[0] = "Laranja"
```

```
# Contar a palavra Maçã na tupla
tp.count("Maçã")
```

```
# Fatiamento
tp[0:2]
```

####**Dicionários**

```
# Para criar um dicionário utilizamos as {}
dc = {"Maçã":20, "Banana":10, "Laranja":15, "Uva":5} 
# Dicionários trabalham com o conceito chave e valor
```

```
dc
```

```
# Acessando o valor de um dicionário através da chave
dc["Maçã"]
```

```
# Atualizar o valor da Maçã no dicionário
dc["Maçã"] = 25
dc
```

```
# Retornar todas as chaves do dicionário
dc.keys()
```

```
# Retornar os valores do dicionário
dc.values()
```

```
# Verificar se já existe uma chave no dicionário e caso não exista, é realizado a inserção
dc.setdefault("Limão", 22)
```

```
dc
```
